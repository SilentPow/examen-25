```python
import pandas as pd

def get_dataframe(name):
    return pd.read_csv(name)
```

```python
customers = get_dataframe('customers.csv')
orders = get_dataframe('orders.csv')
products = get_dataframe('products.csv')

customers.head()
```

```python
def get_best_combinaisons(orders: pd.DataFrame, products):
    new_orders = orders.groupby('OrderID', as_index=False).agg(list)
    new_orders = new_orders.drop(['CustomerID', 'OrderDate', 'Quantity', 'TotalAmount', 'OrderStatus', 'Company'], axis=1)

    return new_orders

get_best_combinaisons(orders, products).head()
```

```python

new_orders = orders.drop(['OrderDate', 'Quantity', 'TotalAmount', 'OrderStatus', 'Company'], axis=1)
new_customers = customers.drop(['Email', 'JoinDate', 'Location'], axis=1)

df = new_orders.merge(new_customers)
df = df.dropna()
df = df.loc[df['CustomerName'].str.startswith('John ')]

for name, group in df.groupby('CustomerName'):
    group = group.drop(['CustomerID', 'OrderID'], axis=1)
    group.reset_index(drop=True, inplace=True)
    group['Count'] = group.groupby('ProductID')['ProductID'].transform('size')
    group = group.drop_duplicates()
    group = group.sort_values('Count', ascending=False)
    group.reset_index(drop=True, inplace=True)

    group.to_csv(name + '.csv')

group.head()

```
